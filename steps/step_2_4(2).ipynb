# Подвиг 4

```python
import numpy as np
import matplotlib.pyplot as plt

def func(x):
    return 0.4 * x + 0.1 * np.sin(2*x) + 0.2 * np.cos(3*x)

def df(x):
    return 0.4 + 0.2 * np.cos(2*x) - 0.6 * np.sin(3*x)

eta = 1.0
N = 500
x_momentum, x_vanilla = 4.0, 4.0
gamma = 0.7
v = 0

for i in range(N):
    v = gamma * v + (1 - gamma) * eta * df(x_momentum - gamma * v)
    x_momentum = x_momentum - v
    x_vanilla = x_vanilla - eta * df(x_vanilla)


X = np.arange(-5, 10, 0.01)
y = func(X)
plt.plot(X, y)
plt.scatter(x_vanilla, func(x_vanilla), color='b', label=f'Method no momentum: {round(x_vanilla, 2)}')
plt.scatter(x_momentum, func(x_momentum), color='r', label=f'Method momentum: {round(x_momentum, 2)}')
plt.grid(True)
plt.legend()
plt.show()
```
