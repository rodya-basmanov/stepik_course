# Подвиг 3

```python
import numpy as np
import matplotlib.pyplot as plt


def func(x):
    return -0.5 * x + 0.2 * x ** 2 - 0.01 * x ** 3 - 0.3 * np.sin(4*x)


def df(x):
    return -0.5 + 0.4 * x - 0.03 * x ** 2 - 1.2 * np.cos(4 * x)


# здесь объявляйте функцию df (производную) и продолжайте программу
eta = 0.1
N = 200
x_momentum, x_vanilla = -3.5, -3.5
gamma = 0.8
v = 0


for _ in range(N):
    v = gamma * v + (1 - gamma) * eta * df(x_momentum)
    x_momentum = x_momentum - v
    x_vanilla = x_vanilla - eta * df(x_vanilla)

X = np.arange(-5, 10, 0.01)
y = func(X)
plt.plot(X, y)
plt.scatter(x_vanilla, func(x_vanilla), color='b', label='Method no momentum')
plt.scatter(x_momentum, func(x_momentum), color='r', label='Method momentum')
plt.grid(True)
plt.legend()
plt.show()
```
